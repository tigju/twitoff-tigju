#### 1. Convert to lower case

```python
data['text_lower']  = data['full_text'].str.lower()
data['text_lower'].head(10)
```

#### 2. remove hyperlinks from tweets

```python
import re

def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+', flags=re.MULTILINE)
    return url_pattern.sub(r'', text)

data['no_links_text'] = data['text_lower'].apply(remove_urls)

data['no_links_text'].head(10)
```

#### 3. Convert emojis and emoticons to words with emot library

```python
from emot.emo_unicode import UNICODE_EMO, EMOTICONS
```

```python
def convert_emoticons(text):
    for emot in EMOTICONS:
        text = re.sub(u'('+emot+')', " ".join(EMOTICONS[emot].replace(",","").split()), text)
    return text
    
def convert_emojis(text):
    for emot in UNICODE_EMO:
        text = text.replace(emot, " ".join(UNICODE_EMO[emot].replace(",","").replace(":","").split()))
    return text
    
data['covn_emot'] = data['no_links_text'].apply(convert_emoticons)
data['covn_emoji'] = data['covn_emot'].apply(convert_emojis)

data['covn_emoji'].head(20)
```

```python
# separate emojis from text ex.: @djkhaleTongue -> @djkhale Tongue
def sep_emoji(text):
    new_text = re.sub('([A-Z])', r' \1', text)
    return new_text.lower()

data['covn_emoji'] = data['covn_emoji'].apply(sep_emoji)

data['covn_emoji'].head(10)
```

#### 3. Tokenize words

```python
from nltk.tokenize import word_tokenize
def doc_to_tokens(doc):
    return word_tokenize(doc)

data['tokenized_docs'] = data['abbr_to_words'].apply(doc_to_tokens)

data['tokenized_docs'].head(10)
```

#### 4. Remove punctuation

```python
import string

def remove_punct(tokenized_docs):
    regex = re.compile('[%s]' % re.escape(string.punctuation))
    tokenized = []
    for token in tokenized_docs:
        new_token = regex.sub(u'', token)
        if new_token != u'':
            tokenized.append(new_token)
        
    return tokenized

data['tokens_no_punct'] = data['tokenized_docs'].apply(remove_punct)
data['tokens_no_punct'].head(10)
```

#### 5. Remove stop words

```python
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('english'))

def remove_stopwords(tokenized_docs):
    
    no_stopwords = []
    for token in tokenized_docs:
        if token not in STOPWORDS:
            no_stopwords.append(token)
        
    return no_stopwords

data['no_stopwords'] = data['tokens_no_punct'].apply(remove_stopwords)
data['no_stopwords'].head(10)
```

#### 5. Spelling correction

```python
# this step takes some tome to process
from textblob import TextBlob

def spelling(doc):
    new_doc = []
    for token in doc:
        new_token = str(TextBlob(token).correct())
        new_doc.append(new_token)
    return new_doc

data['spelling'] = data['no_stopwords'].apply(spelling)
```
